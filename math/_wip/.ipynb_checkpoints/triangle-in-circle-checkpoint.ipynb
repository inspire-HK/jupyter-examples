```python
import numpy as np
import matplotlib.pyplot as plt
import math

theta = np.linspace(0, 2*np.pi, 100)

x1 = np.cos(theta) 
y1 = np.sin(theta) 

plt.plot(x1, y1)

x = np.array([0,120,240,0])
y = np.array([0,120,240,0])
x = np.cos( x * np.pi / 180 )
y = np.sin( y * np.pi / 180 )

plt.plot( x, y, color='red' )
plt.show()
```
